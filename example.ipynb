# Display config

```python
def format_float(float_num):
    return '{:,.2f}'.format(float_num).replace(',', ' ')
```

```python
pd.set_option('max_colwidth', None)
pd.set_option('max_rows', None)
pd.set_option('float_format', format_float)
```

# Transpose data

In many cases, you'll want to use sections in the balance sheet as features. In those circumtances, you might need to transpose data.

In this example, we will plot top 10 companies that have the highest `TNG CNG NGUN VN`

```python
t_bs4 = bs4.T
t_bs4.head()
```

```python
top10 = t_bs4['TNG CNG NGUN VN'].nlargest(10)
top10
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=top10.index, y=top10, palette="Blues_d")
plt.show()
```
